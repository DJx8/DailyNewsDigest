### Load the JSON Dataset

```python
import pandas as pd

# Path to your JSON file (update path if needed)
file_path = '/content/drive/MyDrive/NLP/Project/Dataset/News_Category_Dataset_v3.json'
df = pd.read_json(file_path, lines=True)

# Show first few rows
df.head()
```

```python
category_counts = df['category'].value_counts()
print(category_counts)
```

### Remove Unnecessary / Noisy Categories

```python
# Categories to remove
unnecessary_categories = [
    "CULTURE & ARTS","ARTS & CULTURE","MEDIA","WOMEN","BLACK VOICES", "QUEER VOICES","PARENTING", "PARENTS", "STYLE", "STYLE & BEAUTY",
    "WORLDPOST", "HEALTHY LIVING", "TASTE", "GOOD NEWS", "IMPACT","WEDDINGS","THE WORLDPOST",
    "WEIRD NEWS", "GREEN", "LATINO VOICES", "FIFTY", "DIVORCE","COLLEGE","U.S. NEWS","MONEY","RELIGION"
]

# Filter them out
df = df[~df['category'].isin(unnecessary_categories)]

# Optional: Reset index
df = df.reset_index(drop=True)

# Confirm final category count
df['category'].value_counts()
```

### Preprocess the Data Columns

```python
# Combine headline + short description
df['text'] = df['headline'].fillna('') + ". " + df['short_description'].fillna('')

# Drop rows with empty text or category
df = df[(df['text'].str.strip() != "") & (df['category'].str.strip() != "")]
```
